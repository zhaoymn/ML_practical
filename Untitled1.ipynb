```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pickle as cp
import math
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

from sklearn.datasets import load_iris
class NBC:
    feature_num = 0
    feature_type = 0
    class_num = 0
    train_data_num = 0
    parameters = []
    pB = []
    pA = []
    
    def __init__(self,feature_types,num_classes):
        self.feature_num = len(feature_types)
        self.class_num = num_classes
        self.feature_type = feature_types
        self.parameters = [0 for i in range(self.feature_num)]
        self.pB = [0 for i in range(self.feature_num)]
        self.pA = [0 for i in range(self.class_num)]
    
    def get_gaussian_value(self, x, mean, sigma):
        return 1.0*math.exp(-((x-mean)*(x-mean))/(2*sigma))/(math.sqrt(sigma*2*math.pi))
    
    def fit(self, Xtrain, ytrain):
        self.train_data_num = len(ytrain)
        for i in range(self.class_num):
            self.pA[i] = dict()
            self.pA[i]["count"] = 0
            self.pA[i]["p"] = 0
        for i in range(len(self.feature_type)):
            self.pB[i] = dict()
            self.parameters[i] = dict()
            if self.feature_type[i] == 'b':
                for j in range(len(ytrain)):
                    
                    item = Xtrain[j][i]
                    if item in self.pB[i]:
                        self.pB[i][item]["count"] = self.pB[i][item]["count"] + 1
                    else:
                        self.pB[i][item] = dict()
                        self.pB[i][item]["count"] = 1
                        self.pB[i][item]["p"] = 0
                    
                    if "count" in self.pA[ytrain[j]]:
                        self.pA[ytrain[j]]["count"] = self.pA[ytrain[j]]["count"] + 1
                    else:
                        self.pA[ytrain[j]]["count"] = 1
                        
                    if item in self.parameters[i].keys():
                        self.parameters[i][item][ytrain[j]] = self.parameters[i][item][ytrain[j]] + 1
                        self.parameters[i][item]["total"] = self.parameters[i][item]["total"] +1
                    else:
                        self.parameters[i][item] = dict()
                        self.parameters[i][item]["total"] = 1
                        for k in range(self.class_num):
                            self.parameters[i][item][k] = 0
                        self.parameters[i][item][ytrain[j]] = 1;
                for key in self.pB[i].keys():
                    self.pB[i][key]["p"] = self.pB[i][key]["count"]/self.train_data_num
                for j in range(self.class_num):
                    for item in self.parameters[i].keys():
                        if j in self.parameters[i][item]:
                            self.parameters[i][item][j] = self.parameters[i][item][j]/self.parameters[i][item]["total"]
                        else:
                            self.parameters[i][item][j] = 1e-6
            else:
                for j in range(self.class_num):
                    temparray = []
                    for k in range(len(ytrain)):
                        if ytrain[k] == j:
                            temparray.append(Xtrain[k,i])# = [temparray Xtrain[k,i]]
                    self.parameters[i][j] = dict()
                    self.parameters[i][j]["mean"] = np.mean(temparray)
                    #print(self.parameters[i][j]["mean"]) 
                    #print(type(temparray))
                    varian = np.var(temparray)
                    if varian ==0:
                        varian = 1e-6
                    self.parameters[i][j]["sigma"] = varian
                pass
        for i in range(self.class_num):
            self.pA[i]["p"] = self.pA[i]["count"]/self.train_data_num
        
    
    def predict(self, Xtest):
        result = list()
        for k in range(len(Xtest)):
            probability = [0 for i in range(self.class_num)]
            for i in range(self.class_num):
                for j in range(len(self.feature_type)):
                    if self.feature_type[i] == 'b':
                        if Xtest[k][j] in self.parameters[j]:
                            probability[i] = probability[i] + np.log(self.parameters[j][Xtest[k][j]][i]*self.pA[i]["p"]/(self.pB[j][Xtest[k][j]]["p"]+1e-6)+1e-6)
                        else:
                            probability[i] = probability[i] + np.log(1e-6)
                    else:
                        probability[i] = probability[i] + np.log(self.get_gaussian_value(Xtest[k][j],self.parameters[j][i]["mean"],self.parameters[j][i]["sigma"]))
            bestclass = 0;
            value = probability[0]
            for i in range(self.class_num):
                if probability[i] > value:
                    value = probability[i]
                    bestclass = i
            result.append(bestclass)
        return result

```

```python
#handin 1 the choice of C
'''
In NNL Expression the ||w||^2 term has a coefficient of lambda. 
From http://scikit-learn.org/stable/modules/linear_model.html#logistic-regression we can see that when lambda
equals 0.5 the coefficient of the main term is C. So that C = 0.5/lamba. When lambda = 0.1, C = 5.
'''



#handin 2
#for iris dataset
iris = load_iris()
X, y = iris['data'], iris['target']
N, D = X.shape
Ntrain = int(0.8*N)

result_NBC = [0.0]*10
result_LR = [0.0]*10
runtimes = 100
for i in range(runtimes):
    shuffler = np.random.permutation(N)
    Xtrain = X[shuffler[:Ntrain]]
    ytrain = y[shuffler[:Ntrain]]
    Xtest = X[shuffler[Ntrain:]]
    ytest = y[shuffler[Ntrain:]]
    for k in range(10):
        train_num = round((k+1)*Ntrain/10)
        X_train_new = Xtrain[:train_num]
        y_train_new = ytrain[:train_num]
        nbc = NBC(['r','r','r','r'],4)
        nbc.fit(X_train_new,y_train_new)
        predict = np.array(nbc.predict(Xtest)).T
        error = 1.0 - np.mean(ytest == predict)
        result_NBC[k] = result_NBC[k] + error
        
        LR = LogisticRegression(C=5)
        LR.fit(X_train_new, y_train_new)
        predict = LR.predict(Xtest)
        error = 1.0 - np.mean(ytest == predict)
        result_LR[k] = result_LR[k] + error

for i in range(10):
    result_NBC[i] = result_NBC[i]/runtimes
    result_LR[i] = result_LR[i]/runtimes
print(result_NBC)
print(result_LR)

fig,ax = plt.subplots()
ax.plot(result_NBC)
ax.plot(result_LR)
fig.tight_layout()
plt.show()

#voting dataset
X, y = cp.load(open('voting.pickle','rb'))
y = y.astype(int)
N, D = X.shape
Ntrain = int(0.8*N)

result_NBC = [0.0]*10
result_LR = [0.0]*10
runtimes = 100
for i in range(runtimes):
    shuffler = np.random.permutation(N)
    Xtrain = X[shuffler[:Ntrain]]
    ytrain = y[shuffler[:Ntrain]]
    Xtest = X[shuffler[Ntrain:]]
    ytest = y[shuffler[Ntrain:]]
    for k in range(10):
        train_num = round((k+1)*Ntrain/10)
        X_train_new = Xtrain[:train_num]
        y_train_new = ytrain[:train_num]
        nbc = NBC(['b']*16,2)
        nbc.fit(X_train_new,y_train_new)
        predict = np.array(nbc.predict(Xtest)).T
        error = 1.0 - np.mean(ytest == predict)
        result_NBC[k] = result_NBC[k] + error
        
        LR = LogisticRegression(C=5)
        LR.fit(X_train_new, y_train_new)
        predict = LR.predict(Xtest)
        error = 1.0 - np.mean(ytest == predict)
        result_LR[k] = result_LR[k] + error

for i in range(10):
    result_NBC[i] = result_NBC[i]/runtimes
    result_LR[i] = result_LR[i]/runtimes
print(result_NBC)
print(result_LR)

fig,ax = plt.subplots()
ax.plot(result_NBC)
ax.plot(result_LR)
fig.tight_layout()
plt.show()

```
